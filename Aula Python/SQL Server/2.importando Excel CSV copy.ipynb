```python
#Substituir valores nulos
data['street'] =           data['street'].fillna('Nao Cadasatrado')
data['email'] =            data['email'].fillna('Email Não Cadastrado')
data['number'] =           data['number'].fillna('Sem Numero')
data['additionals'] =      data['additionals'].fillna('Sem Info')
```

#### Metodo Replace

```python
data['state'] = data['state'].replace('São Paulo', 'SP')
data['state'] = data['state'].replace('Acre', 'AC')
data['state'] = data['state'].replace('Distrito Federal', 'DF')
```

```python
data.head()
```

```python
# Fazendo a substituição de um valor específico
valor_original = 'SP'
valor_substituto = 'São Paulo'
data['state'] = data['state'].replace(valor_original, valor_substituto)
```

```python
data.head()
```

```python
# Mapeando os nomes dos estados para substituição
# Obs itens criados entre chves {} significa dicionario
mapeamento = {'Maranhão': 'Maranhao', 'Piauí': 'Piaui', 'Goiás': 'Goias','São Paulo':'Sao Paulo'}

# Substituindo os nomes dos estados no DataFrame
data['state'] = data['state'].replace(mapeamento)
```
