```python
import pandas as pd
data = pd.read_csv(r"C:\Users\TERRA\JAVA\dio-java-basico\udemy1\Origem\arquivos_csv\Clientes.csv", delimiter=',')
```

```python
data['nome completo'] = data['first_name'] + ' ' + data['last_name']
data
```

```python
data.info()
```

```python
# Converter a coluna 'created_at' para o formato de data
data['created_at'] = pd.to_datetime(data['created_at'])

print(data['created_at'].dtype)
```

```python
# Adicionar coluna com o ano
data['ano'] = data['created_at'].dt.year
data
```

```python
# Adicionar coluna com o ano
data['Mes'] = data['created_at'].dt.month
data
```

```python
# Adicionar coluna com o dia
data['dia'] = data['created_at'].dt.day
data
```

```python
# Adicionar coluna com o nome do mês em português
meses = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
         'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']
data['Nomemes'] = data['created_at'].dt.month.map(lambda x: meses[x-1])
data
```

```python
# Adicionar coluna com o dia
data['Hora'] = data['created_at'].dt.hour
data['Minuto'] = data['created_at'].dt.minute
data['Segundo'] = data['created_at'].dt.second
data  
```
