```python
#ordem crescente coluna especifica
dados   =    dados.sort_values(by=['Name'])
dados.head(10)
```

```python
#ordem decrescente coluna especifica
dados=dados.sort_values(by=['Name'],ascending = False )
dados.head(10)
```
